```python
# The first step in data cleaning is to import all libraries needed:

import re
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import argparse
import datetime
import requests
from lxml import html
import lxml.html as lh
from bs4 import BeautifulSoup
import warnings 
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv('df_to_clean')
display(df)
```

```python
df[['work_field', 'company']] = df.Source.str.split(' ==> ', expand=True)
df
```

```python
def drop_cols(lst):
    for i in lst:
        df.drop(i, axis=1, inplace=True)
```

```python
cols_to_drop=['Unnamed: 0', 'id', 'Unnamed: 0_x', 'realTimeWorth', 'Unnamed: 0_y', 'Unnamed: 0.1', 'Source']
drop_cols(cols_to_drop)
```

```python
clean_text = {'worth': ' BUSD', 'worthChange': ' millions USD', 'age': ' years old'}
def clean_numcols_from_text(dic):
    for k, v in dic.items():
        df[k] = df[k].str.replace(v, '')
clean_numcols_from_text(clean_text)
```

```python
def column_title(df, ls):
    for i in ls:
        df.update(df[i].str.title())
    return df
ls_cols = ['name', 'lastName', 'company']
column_title(df, ls_cols)
```

```python
df = df.rename(columns={'name': 'first_name', 'lastName': 'last_name', 'realTimePosition': \
                        'real_time_position', 'worth': 'worth(BUSD)', 'worthChange': 'worth(millions_USD)'})
df = df[['position', 'first_name', 'last_name', 'age', 'gender', 'country', 'real_time_position', \
              'worth(BUSD)', 'worth(millions_USD)', 'work_field', 'company', 'image']]

display(df)
```

```python
# Since we had the first and last name in the column name, with the following code we have only the first
# name in the column name and the last name in the column lastName
df['first_name'] = [a.replace(b, '').strip() for a, b in zip(df['first_name'], df['last_name'])]

display(df)
```

```python
# This command is necessary to check the values we have in this series:
df['gender'].value_counts()
```

```python
df.gender.replace('Male', 'M', inplace=True)
df.gender.replace('Female', 'F', inplace=True)
display(df)
```

```python
# Transform all ages to correct numbers:
df.age.fillna(str(-2019), inplace=True)
current_year = datetime.datetime.today().year

age = []
for i in df['age']:
    if len(str(i)) == 4:
        age.append(str(current_year-int(i)))
    else:
         age.append(i)

df['age'] = age
display(df)
```

```python
# The function change_types receives a list of columns of the df as an argument and will convert its type
# to float, so it's possible to operate with them.
def change_to_float_type(l):
    for i in l:
        df[i] = df[i].astype(float)
change_to_float_type(['worth(BUSD)', 'worth(millions_USD)'])
```

```python
# To complete our data I had to get this file:
path = '../data/raw/forbes_2018.csv'
forbes = pd.read_csv(path)
forbes.tail(15)
```

```python
forbes.dropna(subset=['position'], inplace=True)

forbes.age.isnull().sum()
```

```python
df['gender'] = df['position'].map(forbes.set_index('position')['gender'])
df['country'] = df['position'].map(forbes.set_index('position')['country'])
```

```python
# Since names from china were wrong
count=-1
df['change_name'] = np.nan
for i in df['country']:
    count+=1
    if i == 'China':
        df['change_name'][count]=df['last_name'][count]
        df['last_name'][count]=df['first_name'][count]
        df['first_name'][count] = df['change_name'][count]
    else:
        pass
```

```python
df.drop('change_name', axis=1, inplace=True)
display(df)
```

```python
# The column 'name' is going to be used for wikipedia web scraping and the column education is going to be updated
# by the scraped data.
df['name'] = df['first_name'] + '_' + df['last_name']
df['education'] = 'None'
```
